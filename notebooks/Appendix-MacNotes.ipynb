## Homebrew

apt does not exist on mac. You can install  [Homebrew](https://brew.sh/) which works much the same. For example to install wget or any other command line program we simply type:

```bash
brew search wget

==> Formulae
wget
wgetpaste
```

To search the hombrew apps and see if the app we want is available. and then:
```bash
brew install wget
```
to install the application.

We can also install graphical applications using the `cask` command. For example to install the atom text editor we would first search for the cask in brew:

```
brew search atom

==> Formulae
atomicparsley
datomic
libatomic_ops
sratom

==> Casks
atom
homebrew/cask-versions/atom-beta
```
Note that the available apps are split into the `Formulae` and `Casks` sections. Here the `Formulae` are command line tools which can be installed directly using `brew install` whilst the `Casks` are graphical applications which require the `cask` command (i.e. `brew cask install`)

So to install our atom text editor we can write:
```
brew cask install atom
```

for more information either look at the manual page `man brew` or the [homebrew docs](https://docs.brew.sh/).

Unfortunately we cannot install TeX from homebrew so we need to use a distribution called [macTeX](http://www.tug.org/mactex/index.html). I recommend that you install the smaller basic Tex distribution as the main macTeX contains alot of unnecessary applications.

Once installed we can get extra packages uisng the TeXLive package manager `tlmgr`. Much like brew we can use this to search for packages:
```
tlmgr search package_name
```

and to install them:
```
tlmgr install package_name
```


## Python
Though OSX comes with a default python install it is best to avoid using it as it can cause problems with the operating system. It's better to install a seperate python from brew and use that version. We can do this by using:

```
brew install python
```

which installs python 3.x. We can access this version of python using the `python3` command and install packages using `pip3`.

## Jupyter

As of today (2/10/18) jupyter has some conflicts when using the python kernal. After we install jupyter
```
pip3 install jupyter
```

we need to uninstall ipykernel and reinstall an earlier version.

```
pip3 uninstall ipykernel
pip3 install ipykernel==4.8.0
```

This should allow jupter to work as normal (i.e. running `jupyter notebook` in the correct directory).
